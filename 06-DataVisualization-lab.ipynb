## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

```python
QUERYY = """
SELECT * FROM DevType
"""

df = pd.read_sql_query(QUERYY,conn)
df.head()
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY3 =  """
SELECT *
FROM master
"""
df = pd.read_sql_query(QUERY3, conn)
df.head()
```

```python
df.columns
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
sns.boxplot(df['Age'])
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
sns.scatterplot(y="Age",x="WorkWeekHrs",data=df)
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
sns.scatterplot(x='WorkWeekHrs',y='CodeRevHrs',size='Age',data=df,sizes=(0,150))
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY4 = """
SELECT DatabaseDesireNextYear , COUNT(Respondent) AS Count  
FROM DatabaseDesireNextYear group by DatabaseDesireNextYear order by Count DESC limit 5
"""

# QUERY4="""SELECT * FROM DatabaseDesireNextYear"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY4,conn)
px.pie(df,values='Count',names='DatabaseDesireNextYear',title='Top by 5 Databases')
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY5 =  """
SELECT Avg(WorkWeekHrs) AS WorkWeekHrs, Avg(CodeRevHrs) AS CodeRevHrs,Age
FROM master WHERE Age BETWEEN 30 AND 35 group by Age 
"""
df = pd.read_sql_query(QUERY5, conn)

px.bar(df,x='Age',y=['CodeRevHrs','WorkWeekHrs'])
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY6="""
SELECT Age ,  Avg(ConvertedComp) As ConvertedComp FROM master WHERE Age BETWEEN 45 AND 60 group by Age
"""

df = pd.read_sql_query(QUERY6, conn)

px.line(df,x="Age",y='ConvertedComp')
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY7 = """
SELECT MainBranch, count(*) AS Count FROM master group by MainBranch
"""

df=pd.read_sql_query(QUERY7,conn)

px.histogram(df,y='MainBranch',x="Count")
```
